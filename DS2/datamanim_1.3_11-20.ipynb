## Question 14.
각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
df14 = channel[["channelname", "subcnt", "ct"]].copy()
df14 = df14[(df14["ct"] >= pd.to_datetime("2021-10-03 03:00:00")) \
            & (df14["ct"] <= pd.to_datetime("2021-11-01 15:00:00"))]

list_channel = df14["channelname"].drop_duplicates()
list_delta = list()
for obj in list_channel:
    delta = df14[df14["channelname"] == obj].sort_values("ct")["subcnt"].iloc[-1] \
        - df14[df14["channelname"] == obj].sort_values("ct")["subcnt"].iloc[0]
    list_delta.append(delta)

# df14_1 = pd.DataFrame(list_delta, np.array(list_channel))
# df14_1

tmp_dict = {"channelname": np.array(list_channel), "del": list_delta}
df14_1 = pd.DataFrame(tmp_dict)
df14_1
```

## Question 19
2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

```python
df19 = video.copy()

start = pd.to_datetime("2021-11-01 00:00:00")
end = pd.to_datetime("2021-11-01 15:00:00")

def check_agg(x):
    return max(x) - min(x)

def check_app(x):
    return x.max() - x.min()

df19 = df19[(df19.ct >= start) & (df19.ct <= end)]
# df19[["videoname", "viewcnt"]].groupby("videoname").agg(check_agg)
df19[["videoname", "viewcnt"]].groupby("videoname").apply(check_app)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv")
df.head()
```

```python
'''
1.
인기동영상 제작횟수가 많은 채널 상위 10개명을 출력하라 (날짜기준, 중복포함)
'''
list(df.groupby("channelTitle").count().sort_values("title", ascending=False).head(10).index)
```

```python
'''
2.
논란으로 인기동영상이 된 케이스를 확인하고 싶다. 
dislikes수가 like 수보다 높은 동영상을 제작한 채널을 모두 출력하라
'''
list(df[df["dislikes"] > df["likes"]]["channelTitle"].unique())
```

```python
'''
3.
채널명을 바꾼 케이스가 있는지 확인하고 싶다. 
channelId의 경우 고유값이므로 이를 통해 채널명을 한번이라도 바꾼 채널의 갯수를 구하여라
'''
ans3 = list()
channelId_uniqueness = df.groupby("channelId")["channelTitle"].nunique()
print(type(channelId_uniqueness))
for index, value in zip(channelId_uniqueness.index, channelId_uniqueness):
    if (value == 1):
        continue
    ans3.append(index)

len(ans3)
```

```python
'''
4.
일요일에 인기있었던 영상들중 가장많은 영상 종류(categoryId)는 무엇인가?
'''
df["trending_date_new"] = pd.to_datetime(df["trending_date2"])
# df[df["trending_date_new"].dt.day_name() == "Sunday"].groupby("categoryId").count().sort_values("title", ascending=False).index[0]
# print(df[df["trending_date_new"].dt.day_name() == "Sunday"].categoryId.head())
df[df["trending_date_new"].dt.day_name() == "Sunday"].categoryId.value_counts().index[0]
```

```python
'''
5.
각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라
'''
df5 = df[["trending_date_new", "categoryId"]].copy()
df5_new = pd.DataFrame()
day_list = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

for day in day_list:
    df5_new[day] = df5[df5["trending_date_new"].dt.day_name() == day].groupby("categoryId").size()

df5_new
```

```python
# 5 답안
group = df.groupby([df['trending_date_new'].dt.day_name(),'categoryId'],as_index=False).size()
#group
answer= group.pivot(index='categoryId',columns='trending_date_new')
display(answer)
```

```python
''' 
6. 
댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다. 
viewcount대비 댓글수가 가장 높은 영상을 확인하라 
(view_count값이 0인 경우는 제외한다)
'''
df6 = df[df["view_count"] > 0].copy()
df6["ans6"] = df6["comment_count"]/df6["view_count"]
df6_1 = df6.sort_values("ans6", ascending=False)
df6_1.iloc[0]["title"]
```

```python
''' 
7.
댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.
viewcount대비 댓글수가 가장 낮은 영상을 확인하라 
(view_counts, ratio값이 0인경우는 제외한다.)
'''
df6_2 = df6[df6["ans6"] != 0].copy()
df6_2 = df6_2.sort_values("ans6")
df6_2.iloc[0]["title"]
```

```python
''' 
8.
like 대비 dislike의 수가 가장 적은 영상은 무엇인가? 
(like, dislike 값이 0인경우는 제외한다)
'''
df8 = df[(df["likes"] != 0) & (df["dislikes"] != 0)].copy()
df8["ans8"] = df8["dislikes"] / df8["likes"]
df8.sort_values("ans8").iloc[0]["title"]
```

```python
''' 
9.
가장많은 트렌드 영상을 제작한 채널의 이름은 무엇인가? 
(날짜기준, 중복포함)
'''
print(df.groupby("channelTitle").size().sort_values(ascending=False).index[0])
print(df.channelTitle.value_counts().index[0])
```

```python
''' 
10.
20회(20일)이상 인기동영상 리스트에 포함된 동영상의 숫자는?
'''
# 데이터를 분석할 때는, 고유값(여기서는 channelId)을 항상 붙여서 관리할 것
srs10 = df[["title", "channelId"]].value_counts()
srs10[srs10 >= 20].size
```

```python
#이게정답이긴함
answer= (df[['title','channelId']].value_counts()>=20).sum()
answer
```
